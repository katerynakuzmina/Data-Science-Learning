```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
values = [0, 1, 2, 3, 4, 5, 6, 7]
squares = [value ** 2 for value in values]
squares
```

```python
plt.plot(values, squares)
plt.show()
```

```python
plt.scatter(values, squares)
plt.show()
```
